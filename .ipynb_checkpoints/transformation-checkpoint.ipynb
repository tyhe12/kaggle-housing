```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer
```

```python
train_data = pd.read_csv('./data/train.csv')
y = train_data['SalePrice'].values

def johnson(y):
    gamma, eta, epsilon, lbda = st.johnsonsu.fit(y)
    yt = gamma + eta * np.arcsinh((y-epsilon)/lbda)
    return yt, gamma, eta, epsilon, lbda

def johnson_inverse(y, gamma, eta, epsilon, lbda):
    return lbda*np.sinh((y-gamma)/eta) + epsilon

j_y = johnson(y)
log_y = np.log(y)

plt.figure(1); plt.title('Original SalePrice')
sns.distplot(y, kde=False, fit=st.norm)

plt.figure(2); plt.title('Log SalePrice')
sns.distplot(log_y, kde=False, fit=st.norm)


plt.figure(3); plt.title('Johnson SU SalePrice')
sns.distplot(j_y[0], kde=False, fit=st.norm)
```
